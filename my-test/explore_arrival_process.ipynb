```python
# 给定rate和cv，画出gamma分布图
# np.random.gamma(self.shape, self.scale, size=batch_size)

import numpy as np
import matplotlib.pyplot as plt

# Given rate and coefficient of variation (cv)
rate = 1.0  # Example rate
cv = 0.5    # Example coefficient of variation

# Calculate shape (k) and scale (theta) parameters
shape = 1 / (cv ** 2)
scale = cv ** 2 / rate

# Generate gamma distribution
batch_size = 1000  # Example batch size
intervals = np.random.gamma(shape, scale, size=batch_size)

# Plot the gamma distribution
plt.hist(intervals, bins=50, density=True, alpha=0.6, color='g')
plt.title('Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 计算两个列表的余弦相似度，不使用sklearn
# Calculate cosine similarity between two lists without using sklearn
import numpy as np

def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

# Example lists
a = [1,1,0,0,1,0,0,0,1,1,0,1]
b = [1,1,0,0,1,0,0,0,1,1,0,0]
print(cosine_similarity(a, b)) 
```
